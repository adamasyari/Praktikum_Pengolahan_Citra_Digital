script di atas berfungsi untuk mendapatkan ukuran citra (tinggi dan lebar) dengan bantuan method shape(), ukuran kemudian di simpan ke dalam variabel x dan y. lalu memunculkan hasil tinggi dan lebar sebuah foto.

```python
sumGray = np.zeros(256).astype(int)
```

script di atas merupakan inisialisasi sumGray yang mana berisi array bernilai 0 sebanyak 256 array dengan tipe int. lalu menghitung jumlah masing masing intensitas piksel pada citra dengan meng-nolkan terlebih dahulu dengan menggunakan syntax "np.zeros".

```python
for i in range(x):
    for j in range(y):
        val = gray_img[i,j] #membuat variabel value yang akan menyimpan nilai dari index tertentu dari variabel image
        sumGray[val] += 1 #untuk setiap nilai index yang disimpan, maka nilai variabel sumGray akan ditambah 1 
```

### SOAL 2

##### Lakukanlah konvolusi pada citra dengan zero padding dan tanpa padding dengan kernel berikut
##### ([[0, 0,-1, 1],
#####    [0, 2, 1, -1],
#####    [-1, 1, 2, 0],
#####    [1, -1, 0, 0]])
##### Kemudian tampilkan gambar dan nilai hasil konvolusi tanpa proses clipping, dan dengan proses clipping

```python
kernel = np.array([
    [0, 0, -1, 1],
    [0, 2, 1, -1],
    [-1, 1, 2, 0],
    [1, -1, 0, 0]
])
```

script di atas merupakan inisialisasi variabel kernel yang akan digunakan untuk melakukan proses konvolusi pada citra yang ada.

```python
citraHasil = np.zeros((x - 3, y - 3))
#citra hasil untuk hasil konvolusi, 

for i in range(x-3):
    for j in range(y-3):
        citraHasil[i][j] = gray_img[i][j] * kernel[0][0] + gray_img[i][j + 1] * kernel[0][1] + gray_img[i][j + 2]*kernel[0][2] + gray_img[i][j + 3]*kernel[0][3] + gray_img[i + 1][j]*kernel[1][0] + gray_img[i + 1][j + 1] * kernel[1][1] + gray_img[i + 1][j + 2] * kernel[1][2] + gray_img[i + 1][j + 3] * kernel[1][3] + gray_img[i + 2][j] * kernel[2][0] + gray_img[i + 2][j + 1] * kernel[2][1] + gray_img[i + 2][j + 2] * kernel[2, 2] + gray_img[i + 2][j + 3] * kernel[2][3] + gray_img[i + 3][j] * kernel[3][0] + gray_img[i + 3][j + 1] * kernel[3][1] + gray_img[i + 3][j + 2] * kernel[3, 2] + gray_img[i + 3][j + 3] * kernel[3][3]
plt.imshow(citraHasil, cmap = 'gray')
plt.title('Konvolusi tanpa padding tanpa clipping')
```

```python
citraHasil.shape
```

block code di atas berfungsi untuk melakukan proses pengolahan citra untuk dilakukan operasi konvolusi citra tanpa padding dan tanpa clipping. lalu di akhir line code diatas akan dilakukan pemanggilan fungsi imshow dari library matplotlib yang berfungsi untuk menampilkan citra yang telah diolah ke layar.

```python
#clipping adalah cara untuk menormalisasi setiap pixel yang tidak ada ada pada range
#dengan langsung di sederhanakan menjadi minimal range (pada citra ini 8 bit yaitu 0)
#atau maximal range (pada citra ini 8 bit yaitu 255)
for i in range(x-3):
    for j in range(y-3):
        if citraHasil[i][j] < 0:
            citraHasil[i][j] = 0
        if citraHasil[i][j] > 255:
            citraHasil[i][j] = 255
plt.imshow(citraHasil, cmap  = 'gray')
plt.title('Konvolusi Tanpa padding dengan Clipping')
```

block code di atas berfungsi untuk menampilkan hasil konvolusi dengan clipping ke layar menggunakan syntax imshow dari library matplotlib

```python
result = np.zeros((x, y))
padding_image = np.pad(gray_img, [(1, 2), (1, 2)])
for i in range(x):
    for j in range(y):
        x = i + 1
        y = j + 1
        for k in range(-1, 3):
            for l in range(-1, 3):
                result[i][j] += padding_image[x + k][y + l] * kernel[k + 1][l + 1]
plt.imshow(result, cmap = "gray")
plt.title('Konvolusi dengan padding tanpa clipping')
print(padding_image)
```

block code di atas berfungsi untuk melakukan proses padding citra dan akan dilakukan operasi konvolusi setelah di padding proses diatas dilakukan tanpa clipping. setelah itu hasil konvolusi dengan padding dan tanpa clipping akan ditampilkan ke layar dengan syntax imshow dari library matplotlib

```python
for i in range(x):
    for j in range(y):
        if(result[i][j] < 0):
            result[i][j] = 0
        if(result[i][j] > 255):
            result[i][j] = 255

plt.imshow(result, cmap = "gray")
plt.title('Konvolusi dengan padding dengan clipping')
print(result)
```

block code di atas berfungsi untuk melakukan proses clipping setelah sebelumnya dilakukan proses konvolusi dengan padding. selanjutnya hasil proses clipping akan ditampilkan ke layar dengan syntax imshow dari library matplotlib
