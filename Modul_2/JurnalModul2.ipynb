### SOAL NOMOR 1
#### Lakukan operasi dibawah ini menggunakan gambar diatas:

```python
image = plt.imread("beanny.jpg")
plt.imshow(image)
```

##### Script di atas berfungsi untuk memasukkan gambar ke dalam codingan 

```python
grayscale_image = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
plt.imshow(grayscale_image, cmap = "gray")
```

##### Script di atas berfungsi untuk mengubah warna gambar berwarna menjadi warna putih dan abu menggunakan script "cv.COLOR_RGB2GRAY"

#### a. translasi sejauh x = -50 dan y = 40

```python
panjang, lebar = grayscale_image.shape 
print(panjang, lebar)
```

##### Script di atas berfungsi untuk mendeklarasikan panjang dan lebar pada gambar 

```python
x = -50
y =  40

translasi = np.zeros((panjang, lebar)).astype(int)

for i in range(panjang):
    for j in range(lebar):
        if i - 40 > 0 and j + 50 < lebar:
            translasi[i][j] = grayscale_image[i - 40][j + 50]
```

##### Script di atas berfungsi untuk transalasi yang dimana translasi dapat dilakukan dengan cara menggeser setiap piksel pada citra sejauh nilai translasi yang ditentukan pada sumbu x dan y. Dengan demikian, hasil translasi citra akan memiliki perbedaan posisi piksel yang ditentukan pada sumbu x dan y dengan citra asli.

```python
plt.subplot(121)
plt.imshow(grayscale_image, cmap = "gray")
plt.subplot(122)
plt.imshow(translasi, cmap = "gray")
plt.show()
```

##### Script di atas berfungsi untuk menampilkan hasil proses dari transalsi gambar.

#### b. Rotasi dengan sudut 70 derajat

```python
rotation = np.zeros((panjang, lebar)).astype(int)
sudut = np.deg2rad(70)
midRow = int((panjang) / 2)
midCol = int((lebar) / 2)

for i in range (panjang):
    for j in range (lebar):
        y = (i - midCol) * np.cos(sudut) + (j - midRow) * np.sin(sudut)
        x = -(i - midCol) * np.sin(sudut) + (j - midRow) * np.cos(sudut)
        y = int(y + midCol)
        x = int(x + midRow)
        if x >= 0 and y >= 0 and x < panjang and y < lebar:
            rotation[i][j] = grayscale_image[y][x]
```

##### Script di atas berfungsi untuk merotasikan sebuah gambar yang dimana Rotasi dapat dilakukan dalam arah jarum jam atau berlawanan arah jarum jam dan dapat dilakukan dengan berbagai sudut rotasi.

```python
plt.subplot(121)
plt.imshow(grayscale_image, cmap = "gray")
plt.subplot(122)
plt.imshow(rotation, cmap = "gray")
plt.show()
```

#### c. pencerminan terhadap sumbu x dan sumbu y

```python
cerminX = np.zeros(grayscale_image.shape)
mirror = panjang - 1

for i in range (panjang):
    for j in range (lebar):
        cerminX[i][j] = grayscale_image[mirror][j]
    mirror -= 1
```

##### Script di atas berfungsi untuk pencerminan sumbu X yang dimana Pencerminan pada sumbu X dilakukan dengan cara mengganti baris piksel pada citra dengan urutan yang dibalik dari baris asli. Dalam proses ini, piksel pada baris teratas ditukar dengan piksel pada baris terbawah, piksel pada baris kedua dari atas ditukar dengan piksel pada baris kedua dari bawah, dan seterusnya.

```python
cerminY = np.zeros(grayscale_image.shape)
mirror = panjang - 1

for i in range(panjang):
    for j in range(lebar):
        cerminY[j][i] = grayscale_image[j][mirror]
    mirror -= 1
```

##### Script di atas berfungsi untuk pencerminan sumbu Y yang dimana pencerminan pada sumbu Y dilakukan dengan cara mengganti kolom piksel pada citra dengan urutan yang dibalik dari kolom asli. Dalam proses ini, piksel pada kolom terkiri ditukar dengan piksel pada kolom terkanan, piksel pada kolom kedua dari kiri ditukar dengan piksel pada kolom kedua dari kanan, dan seterusnya.

```python
plt.subplot(121)
plt.title("Cermin X")
plt.imshow(cerminX, cmap = "gray")
plt.subplot(122)
plt.title("Cermin Y")
plt.imshow(cerminY, cmap = "gray")
plt.show()
```

##### Script di atas berfungsi untuk menampilkan hasil gambar dari cermin X dan Y.

#### d. Dilatasi matriks dengan faktor skala = 3


```python
skala = 3
length = panjang * skala
width = lebar * skala
dilatasi = np.zeros((length, width))

for i in range (lebar):
    for j in range (panjang):
        for k in range (skala):
            dilatasi[i * skala][j * skala + k] = grayscale_image[i][j]
            dilatasi [i * skala + 1][j * skala + k] = grayscale_image[i][j]
```

#### Script di atas berfungsi untuk dilatasi dengan faktor skala pada citra digital dapat digunakan untuk memperbesar citra dengan mempertahankan kualitas citra yang lebih baik, tergantung pada teknik interpolasi yang digunakan.

```python
plt.imshow(dilatasi, cmap = "gray")
```

##### Script di atas berfungsi untuk menampilkan hasil gambar yang sudah dilatasi dengan skala = 3.

### SOAL NOMOR 2 
#### Lakukan Operasi Citra Negatif dari citra grayscale lalu bandingkan histogram kedua citra tersebut.

```python
negative = 255 - grayscale_image

plt.subplot(121)
plt.imshow(grayscale_image, cmap = "gray")
plt.subplot(122)
plt.imshow(negative, cmap = "gray")
plt.show()
```

##### Script di atas berfungsi untuk mendapatlan citra negatif dengan cara mengurangi nilai instensitas pixel dari nilai keabuan maksimum, lalu ditampilkan hasil dari pengurangan intensitas pixel tersebut.

```python
gray_image = grayscale_image
```

##### Script di atas berfungsi unutk mendeklasarasikan gray_image.

```python
derajat = np.zeros(256)
for i in range (256):
    derajat[i] = i
```

##### Script di atas berfungsi untuk menampilkan derajat keabuan dan derajat negatif yang dimana dilakukannya iterasi dengan nilai intensitas pixel pada citra.

```python
in_gray = np.zeros(256)
for row in gray_image:
    for col in row:
        in_gray[col] += 1
```

##### Script di atas berfungsi untuk proses nested lopping yang dimana array in_gray digunakan untuk menampung penambahan nilai intensitas pixel dari gambar keabuan. 

```python
in_negative = np.zeros(256)
for row in negative:
    for col in row:
        in_negative[col] += 1
```

##### Script di atas berfungsi untuk proses nested lopping yang dimana array in_negative digunakan untuk menampung penambahan nilai intensitas pixel dari gambar negatif.

```python
plt.xlabel("Biru = Asli, Orange = Negatif")
plt.bar(derajat, in_gray) 
plt.bar(derajat, in_negative)
```

##### Script di atas berfungsi untuk menampilkan histogram gambar keabuan dan histogram gambar negatif
