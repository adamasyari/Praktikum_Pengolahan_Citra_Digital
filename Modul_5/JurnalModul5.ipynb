Selanjutnya, citra grayscale tersebut ditampilkan menggunakan fungsi histogram.imshow() dengan parameter cmap="gray", yang menunjukkan bahwa tampilan citra yang ditampilkan dalam skala abu-abu (grayscale).

Fungsi

```python
def Median (matriks) :
    array = np.sort (matriks, axis = None)
    total = len (array)
    return (array [int ((total - 1) / 2)])

def Medianfilter(image, horizontal, vertikal):
    tinggi, lebar = image.shape
    hasil = np.zeros(image.shape)
    horizontal2, vertikal2 = int ((horizontal - 1) / 2), int ((vertikal - 1) / 2)
    pad = np.pad (image, ((horizontal2 , horizontal - horizontal2 - 1), (vertikal2, vertikal - vertikal2 - 1)), "edge" )
    for i in range (tinggi):
        for j in range (lebar):
            hasil[i, j] = Median(pad[i : i + horizontal, j : j + vertikal])
    return hasil
```

Code daiatas digunakan untuk ,menghitung nilai median suatu matriks def Median(matriks) pendeklarasi fungsi dengan nama Median dan parameter matriks yang akan digunakan untuk memasukkan matriks input. Selanjutnya array = np.sort(matriks, axis=None) digunakan untuk mengurutkan elemen dalam matriks secara ascending dan menempatkannya dalam array baru yang disebut array. Selanjutnya axis=None menandakan bahwa semua elemen akan diurutkan dalam satu dimensi. Code total = len(array) digunakan untuk menghitung jumlah total elemen dalam array dan menyimpannya dalam variabel total, return (array[int((total-1)/2)]) digunakan untuk menghitung median dari array. Code int((total-1)/2) akan menghasilkan indeks median dalam array. Selanjutnya, kode array[int((total-1)/2)] akan mengambil nilai median dari array dan mengembalikannya sebagai output dari fungsi Median. Selanjutnya def Medianfilter(image, horizontal, vertikal) deklarasi fungsi dengan nama Medianfilter dan tiga parameter, yaitu image (yang merupakan citra grayscale yang akan di-filter), horizontal (yang merupakan ukuran kernel filter median pada sumbu horizontal) dan vertikal (yang merupakan ukuran kernel filter median pada sumbu vertikal), tinggi, lebar = image.shape digunakan untuk mengambil dimensi tinggi dan lebar citra image menggunakan fungsi .shape dan menyimpannya dalam variabel tinggi dan lebar, hasil = np.zeros(image.shape) digunakan untuk membuat array kosong yang ukurannya sama dengan image dan menyimpannya dalam variabel hasil. Selanjutnya horizontal2, vertikal2 = int ((horizontal-1)/2), int ((vertikal-1)/2) digunakan untuk menghitung setengah ukuran dari kernel filter pada sumbu horizontal dan vertikal dan menyimpannya dalam variabel horizontal2 dan vertikal2. Angka setengah tersebut digunakan untuk menentukan batas-batas pada saat melakukan padding pada citra, pad = np.pad(image, ((horizontal2,horizontal-horizontal2-1),(vertikal2,vertikal-vertikal2-1)) , "edge" ) untuk menambahkan padding pada citra image dengan ukuran sesuai dengan ukuran kernel filter dan tipe padding "edge". Padding "edge" menambahkan nilai piksel yang terletak pada tepi citra pada bagian padding sehingga nilai piksel di sekitar tepi citra akan disamakan dengan nilai piksel di tepi citra. Selanjutnya for i in range(tinggi)digunakan untuk melakukan loop untuk setiap baris pada citra image. Selanjutnya for j in range(lebar) code ini akan melakukan loop untuk setiap kolom pada citra image. Selanjutnya hasil[i,j] = Median(pad[i:i+horizontal, j:j+vertikal]) digunakan untuk mengambil bagian citra yang sesuai dengan ukuran kernel filter pada setiap posisi pada citra image dan melakukan proses filter median dengan memanggil fungsi Median dan menyimpan hasilnya pada variabel hasil, return hasil digunakan untuk mengembalikan citra hasil filter median sebagai output dari fungsi Medianfilter.

```python
def Mean (matriks) :
    array = np.sum (matriks)
    horizontal, vertikal = matriks.shape
    return (array / (horizontal * vertikal))

def Meanfilter(image, horizontal, vertikal):
    tinggi, lebar = image.shape
    hasil = np.zeros(image.shape)
    horizontal2, vertikal2 = int ((horizontal - 1) / 2), int ((vertikal - 1) / 2)
    pad = np.pad (image, ((horizontal2, horizontal - horizontal2 - 1), (vertikal2, vertikal - vertikal2 - 1)), "edge" )
    for i in range (tinggi):
        for j in range (lebar):
            hasil[i, j] = Mean(pad[i : i + horizontal, j : j + vertikal])
    return hasil
```

Code di atas adalah implementasi dari filter mean atau rata-rata pada gambar/grayscale image. Fungsi Mean menerima input sebuah matriks dan mengembalikan rata-rata dari semua elemen dalam matriks tersebut. Fungsi Meanfilter mengambil gambar, tinggi dan lebar filter horizontal dan vertikal sebagai input, lalu melakukan zero-padding pada gambar menggunakan mode "edge" dan melakukan filter mean pada setiap pixel dalam gambar. Kemudian, hasil filter dikeluarkan sebagai gambar baru.

```python
def Modus(matriks):
    array = np.sort (matriks, axis=None)
    indeks = np.zeros (np.max(array) + 1, dtype = int )
    for i in range (len(array)):
        indeks[array[i]] += 1
        
    cek = np.max(indeks)
    for j in range (len(indeks)):
        if (indeks[j] == cek):
            return j
        
def Modusfilter(image, horizontal, vertikal):
    tinggi,lebar = image.shape
    hasil = np.zeros(image.shape)
    horizontal2, vertikal2 = int ((horizontal - 1) / 2), int ((vertikal - 1) / 2)
    pad = np.pad (image, ((horizontal2, horizontal - horizontal2 - 1), (vertikal2 , vertikal - vertikal2 - 1)), "edge" )
    for i in range (tinggi):
        for j in range (lebar):
            hasil[i, j] = Modus(pad[i : i + horizontal, j : j + vertikal])
    return hasil
```

Code di atas adalah implementasi dari filter modus pada gambar/grayscale image. Fungsi Modus menerima input sebuah matriks dan mengembalikan modus (nilai yang paling sering muncul) dari semua elemen dalam matriks tersebut. Fungsi Modusfilter mengambil gambar, tinggi dan lebar filter horizontal dan vertikal sebagai input, lalu melakukan zero-padding pada gambar menggunakan mode "edge" dan melakukan filter modus pada setiap pixel dalam gambar. Kemudian, hasil filter dikeluarkan sebagai gambar baru.
