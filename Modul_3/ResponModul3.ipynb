## 1. Tampilkan histogram dari citra 1

```python
plt.subplot(121)
plt.title("Hasil Grayscale Citra 1")
citra1 = cv.cvtColor(marc, cv.COLOR_BGR2GRAY)
plt.imshow(citra1, cmap = "gray")

plt.subplot(122)
plt.title("Hasil GrayScale Citra 2")
citra2 = cv.cvtColor(haechan, cv.COLOR_BGR2GRAY)
plt.imshow(citra2, cmap = "gray")
plt.show()
```

```python
grayBar = [i for i in range(256)]
kolom, baris = citra1.shape

def histogram(gambar):
    newGray = np.zeros(256).astype(int)
    
    for i in range(kolom):
        for j in range(baris):
            hasil = gambar[i, j]
            newGray[hasil] += 1
    return newGray

plt.bar(grayBar,histogram(citra1))
```

```python
kolom, baris = citra1.shape
sumGray = histogram(citra1)
Prob = np.zeros(256)

for i in range(256):
    Prob[i] = sumGray[i] / (kolom * baris)

newBar = np.zeros(256)
newBar[0] = Prob[0]

for i in range(1, 256):
    newBar[i] = newBar[i - 1] + Prob[i]

newGrayPixel = citra1.copy()
for i in range(kolom):
    for j in range(baris):
        pixelGambar = citra1[i, j]
        newPixelGambar = round(newBar[pixelGambar] * 255)
        newGrayPixel[i, j] = newPixelGambar

plt.bar(grayBar, histogram(newGrayPixel))
```
